```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/SuperStoreOrders - SuperStoreOrders.csv')
df.head()
```

```python
df.isnull().sum()
```

**Total sales profit Quantity **

```python
df[['sales', 'profit', 'quantity']].sum()
```

```python
df[['sales', 'quantity']].sum()
```

**Category Wise Analysis**

```python
df.groupby('category')['sales'].sum().sort_values(ascending=False)
```

**Category Wise Profit**

```python
df.groupby('category')['profit'].sum()
```

**Graph**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='category', y='sales', data=df)
plt.show()
```

**Sub-Category Analysis**

```python
df.groupby('sub_category')['sales'].sum().sort_values(ascending=False)
df.groupby('sub_category')['profit'].sum().sort_values(ascending=False)
```

```python
sub_profit = df.groupby('sub_category')['profit'].sum().sort_values(ascending=False)
sub_profit
```

```python
import matplotlib.pyplot as plt

# Filter out negative profit values for the pie chart
positive_profit = sub_profit[sub_profit > 0]

plt.figure(figsize=(8,8))
plt.pie(positive_profit, labels=positive_profit.index, autopct='%1.1f%%')
plt.title("Profit by Sub-Category (Positive Profits Only)")
plt.show()
```
