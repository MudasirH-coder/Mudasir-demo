**Importing Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

**Importing the Boston House price Dataset**

```python
house_price_dataset = sklearn.datasets.load_diabetes()
```

```python
print(house_price_dataset)
```

```python
#loading the dataset to a pandas Dataframe
house_price_dataframe = pd .DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
#print First 5 rows of our Dataframe
house_price_dataframe.head()
```

```python
#add the target (price) column to the dataframe
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking the number of rows and columns in the data frame
house_price_dataframe.shape
```

```python
#check the missing valuses
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of the dataset
house_price_dataframe.describe()
```

Understanding the correlation between various features in the dataset


Positive Correlation,
Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
#Constructing a heatmap to understand the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

**Spliting the data and target**

```python
X = house_price_dataframe.drop(['price'], axis=1)
x = house_price_dataframe['price']
```

```python
print(x)
# print(y) # Commented out the erroneous line
```

Spliting the data into training data and testing **data** **bold text**

```python
X_training, X_test, Y_training, Y_test = train_test_split(X,x,test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_training.shape, X_test.shape)
```

**Model Training **

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#traing the model with x_train
model.fit(X_training, Y_training)
```

Evalutation

Prediction on training data

```python
#accuracy for prediction on training data
training_data_predictionv = model.predict(X_training)
```

```python
print(training_data_predictionv)
```

```python
# R Square error
score_1 = metrics.r2_score(Y_training, training_data_predictionv)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_training, training_data_predictionv)

print('R squared error: ', score_1)
print('Mean Absolute Error: ', score_2)
```

Visualizing the actual price and predict prices

```python
plt.scatter(Y_training,  training_data_predictionv)
plt.xlabel("Actual price")
plt.ylabel("predicted prices")
plt.title("Actual price vs predicted prices")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on test data
test_data_predictionv = model.predict(X_test)
```

```python
 #R Square error
score_1 = metrics.r2_score(Y_test, test_data_predictionv)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_predictionv)

print('R squared error: ', score_1)
print('Mean Absolute Error: ', score_2)
```
